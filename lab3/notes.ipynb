```python
import sympy as sp
from math import *
import numpy as np
import matplotlib.pyplot as plt
```

```python
funkcja = lambda f,x: eval(f,globals(), locals())
fx = '2*x**2-4*tan(x)+e**x'

x=sp.Symbol('x')
pochodna = str(sp.diff(fx,x))
x0 = 1.1

print('Wartość x0=',x0)
print('Funkcja=',fx)
print('Wartośc funkcji dla x0=', funkcja(fx,x0))
print('Pochodna funkcji i=', pochodna)
print('Wartość pochodnej dla x0=', funkcja(pochodna,x0))
```

```python
def calka(x,y):
    suma = 0
    for i in range(0,len(x)-1):
        sum = suma + ((y[i+1] +y[i]/2)*(x[i+1]-x[i]))
    return suma
N=10
t=np.arange(0,N,1.0)
s=np.arange(0,N,1.0)

for i in range(0,len(s)):
    s[i] = np.random.rand()

line = plt.plot(t,s,lw=2)
plt.ylim(0,1)
plt.title('Calka='+str(f'{calka(t,s):.2f}'))
plt.show()
```

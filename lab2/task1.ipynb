```python
# Zadanie 1

import matplotlib.pyplot as plt
import numpy as np

def mantysa(bity):
    epsilon = 1.0
    while 1.0 + epsilon != 1.0:
    #while (1.0 + 0.5 * epsilon) != 1.0:
        epsilon = 0.5*epsilon
    return epsilon * (2**bity)

def wykres(x,y,nazwa_y,tytul,skalalog):
    plt.plot(x,y)
    plt.grid(True)
    plt.xlabel('Liczba bitw')
    plt.ylabel(f'{nazwa_y}')
    plt.title(f'{tytul}')
    if skalalog == True:
        plt.yscale("log")
    plt.show()
    
mashine_epsilon = []
mantysa_range = np.arange(8,53)
for i in mantysa_range:
    mashine_epsilon.append(mantysa(i))

wykres(mantysa_range,mashine_epsilon,'Epsilon maszynowy','Epsilon maszynowy od liczby bitw',False)
wykres(mantysa_range,mashine_epsilon,'Skala logarytmiczna','Zlogarytmowany epsilon od liczby bitw',True)
```

```python
def epislon_maszynowy(bits):
    return 2.0**(-bits)
bits_range = np.arange(8, 53)  
epsilon_values =[] 
for m in bits_range:
    epsilon_values.append(epislon_maszynowy(m)) 

wykres(bits_range, epsilon_values,'Epsilon maszynowy','Aproksymacja',True)
```

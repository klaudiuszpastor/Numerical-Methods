```python
# Zadanie 1

import matplotlib.pyplot as plt
import numpy as np

def mantysa(bity):
    epsilon = 1.0
    while 1.0 + epsilon != 1.0:
    #while (1.0 + 0.5 * epsilon) != 1.0:
        epsilon = 0.5*epsilon
    return epsilon * (2**bity)

def wykres(x,y,nazwa_y,tytul,skalalog):
    plt.plot(x,y)
    plt.grid(True)
    plt.xlabel('Liczba bitów')
    plt.ylabel(f'{nazwa_y}')
    plt.title(f'{tytul}')
    if skalalog == True:
        plt.yscale("log")
    plt.show()
    
mashine_epsilon = []
mantysa_range = np.arange(8,53)
for i in mantysa_range:
    mashine_epsilon.append(mantysa(i))

wykres(mantysa_range,mashine_epsilon,'Epsilon maszynowy','Epsilon maszynowy od liczby bitów',False)
wykres(mantysa_range,mashine_epsilon,'Skala logarytmiczna','Zlogarytmowany epsilon od liczby bitów',True)
```

```python
def epislon_maszynowy(bits):
    return 2.0**(-bits)
bits_range = np.arange(8, 53)  
epsilon_values =[] 
for m in bits_range:
    epsilon_values.append(epislon_maszynowy(m)) 

wykres(bits_range, epsilon_values,'Epsilon maszynowy','Aproksymacja',True)
```

Jest to największa liczba nieujemna, której dodanie do jedności daje wynik równy 1. Innymi słowy, 1 + ε = 1 i żadna liczba większa od ε nie spełnia już tego warunku. (Zwróćmy uwagę, że 0,0 też jest taką liczbą, ale w realnych arytmetykach zmiennoprzecinkowych – nie jedyną.)
